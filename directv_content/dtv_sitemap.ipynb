```python
#@title Combine city urls dir_2 and marketing urls dir_1 and dir_2
import pandas as pd
topics = pd.concat([city_url_groups["dir_2"], marketing_url_groups["dir_1"], marketing_url_groups["dir_2"]]).reset_index(drop=True)
topics = topics.dropna(how="all")
topics
```

```python
topics_df = pd.DataFrame({"url_topic":topics})
topics_df["topic_words"] = topics_df["url_topic"].str.replace("-", " ").str.replace("_", " ")
topics_df
# topics_df.to_csv(r"./dtv_url_topics.csv")
```

```python
#@title Identify Most Common Words/Topics in DirecTV's Website
from collections import Counter
url_words = Counter(" ".join(topics_df["topic_words"]).split()).most_common(50)
url_words
```

```python
url_words_df = pd.DataFrame(url_words, columns=['word', 'count'])
url_words_df
```
