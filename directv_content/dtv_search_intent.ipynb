```python
import pandas as pd
intent_kwds = pd.read_csv(r"D:\Users\jackb_HDD\Development\Agency_SEO\data\Sport Team Keyword Expansion - padres_searchers.csv")
intent_kwds
```

```python
from collections import Counter
intent_words = Counter(" ".join(intent_kwds["Keyword"]).split()).most_common(50)
intent_words
```

```python
intent_words_df = pd.DataFrame(intent_words, columns=['word', 'count'])
neg_kwds = ["san", "diego", "padres", "baseball", "game"]
intent_words_df = intent_words_df[~intent_words_df['word'].str.contains('|'.join(neg_kwds))]
intent_words_df.set_index("word", inplace=True)
intent_words_df # ignoring 
```

```python
intent_words_df.iloc[:14,:].plot(kind="bar", figsize=(20,10),alpha=0.75, rot=30, title="Most Frequent Words Associated with a San Diego Padres' Game (Excluding Keyword Inputs)")
```
